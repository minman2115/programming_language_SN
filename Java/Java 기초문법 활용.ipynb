클래스를 이용해서 위의 코드를 개선해보자.

객체(클래스)는 서로 연관된 변수와 메소드를 grouping한 것이다. 그리고 거기에 이름을 붙인것이다.

위의 코드는 회계와 관련된 코드로만 되어있지만 소스중간에 회계가 아닌 다른 주제의 코드내용이 들어오게 된다면 클래스를 이용하면 유용하게 정리할 수 있다.
